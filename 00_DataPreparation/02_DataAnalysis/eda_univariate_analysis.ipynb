```python
import pandas as pd

data = pd.read_csv('../../Database/titanic.csv')
data.head(3)
```

### Categorical Variables

```python
import seaborn as sns
sns.countplot(x='Sex', data=data)
```

```python
print(data['Survived'].value_counts())
sns.countplot(x='Survived', data=data)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))


sns.countplot(x='Survived', data=data, color='orange', ax=axes[0, 0])
sns.countplot(x='Pclass', data=data, ax=axes[0, 1])
sns.countplot(x='Embarked', data=data, color='green', ax=axes[1, 0])
sns.histplot(x='Age', data=data, hue='Survived', kde=True, ax=axes[1, 1])
```
